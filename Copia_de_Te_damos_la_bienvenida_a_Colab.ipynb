**1. ¿Qué es NumPy?**

NumPy (Numerical Python) es una librería fundamental de Python para computación científica. Proporciona un objeto de arreglo multidimensional de alto rendimiento llamado ndarray, junto con funciones matemáticas, estadísticas, algebraicas y de transformación que operan sobre esos arreglos de manera eficiente. Es la base sobre la que se construyen casi todas las demás librerías científicas del ecosistema Python.

**2. ¿Por qué NumPy es más eficiente que las listas tradicionales de Python?**

Las listas de Python almacenan elementos como objetos individuales en memoria (con overhead por cada elemento), mientras que NumPy almacena los datos en bloques contiguos de memoria con un tipo de dato fijo (homogéneo). Esto permite que las operaciones se ejecuten en C compilado en lugar de Python interpretado, lo que puede ser entre 10 y 100 veces más rápido. Además, NumPy aplica operaciones vectorizadas que evitan los bucles explícitos en Python.

**3. ¿Qué es un ndarray?**

Un ndarray (N-dimensional array) es la estructura de datos central de NumPy. Es un arreglo multidimensional homogéneo, es decir, todos sus elementos son del mismo tipo de dato. Puede representar vectores (1D), matrices (2D) o tensores de mayor dimensión. Tiene atributos como shape (forma), dtype (tipo de dato), ndim (número de dimensiones) y size (número total de elementos).

**4. ¿Qué relación existe entre NumPy y el álgebra lineal?**

La relación entre NumPy y el álgebra lineal es muy estrecha, ya que la biblioteca incorpora numerosas funciones para trabajar con vectores y matrices. Entre estas operaciones se encuentran la multiplicación matricial, el cálculo de determinantes, la obtención de matrices inversas y el análisis de valores propios. Estas herramientas son fundamentales para el desarrollo de algoritmos de Inteligencia Artificial y Aprendizaje Automático.

**5. ¿Cómo utilizan los modelos de Deep Learning las matrices de NumPy?**

Los modelos de Deep Learning utilizan matrices para representar datos de entrada, pesos, sesgos y resultados intermedios dentro de las redes neuronales. NumPy facilita el almacenamiento de los datos de entrenamiento, la ejecución de multiplicaciones matriciales y el procesamiento eficiente de operaciones matemáticas que permiten el aprendizaje de los modelos. Gracias a estas capacidades, constituye una de las bases matemáticas sobre las cuales se construyen las redes neuronales modernas.

**6. ¿Qué es Matplotlib?**

Matplotlib es una biblioteca de Python diseñada para la creación de gráficos y visualizaciones de datos. Permite representar información mediante diferentes tipos de gráficos que facilitan el análisis y la interpretación de resultados. Es una de las herramientas más utilizadas en Ciencia de Datos debido a su versatilidad y facilidad de uso.

**7. ¿Qué tipos de gráficos permite construir?**

Matplotlib permite construir gráficos de líneas, gráficos de barras, histogramas, diagramas de dispersión, diagramas circulares, diagramas de caja y otros tipos de visualizaciones estadísticas. Estos gráficos ayudan a comprender mejor los datos y a identificar tendencias, comportamientos y patrones relevantes.

**8. ¿Por qué la visualización de datos es importante en Inteligencia Artificial?**

La visualización de datos es importante porque permite comprender el comportamiento de la información antes de entrenar un modelo de Inteligencia Artificial. A través de gráficos es posible detectar tendencias, anomalías, relaciones entre variables y posibles errores en los datos. Esto facilita la toma de decisiones y mejora la calidad de los análisis realizados.

**9. ¿Cómo ayuda Matplotlib en el Análisis Exploratorio de Datos (EDA)?**

Matplotlib ayuda en el Análisis Exploratorio de Datos permitiendo representar visualmente la información para comprender su comportamiento. **10. ¿Qué es Seaborn?**

Seaborn es una biblioteca de visualización estadística construida sobre Matplotlib. Su objetivo principal es facilitar la creación de gráficos más atractivos y especializados para el análisis estadístico. #**1. GENERACIÓN DE DATOS SINTÉTICOS**


Debido a que no se dispone de información real de una empresa, se construyó un conjunto de datos sintéticos utilizando NumPy. Los datos fueron generados considerando condiciones que pueden encontrarse en una industria real, donde el consumo energético depende principalmente de la producción, las horas de operación y factores ambientales como la temperatura.


```

# Importamos la librería NumPy para realizar cálculos numéricos
import numpy as np

# Importamos Pandas para crear y manipular tablas de datos
import pandas as pd

# Importamos Matplotlib para la creación de gráficos
import matplotlib.pyplot as plt

# Importamos Seaborn para visualizaciones estadísticas
import seaborn as sns


# CONFIGURACIÓN INICIAL

# Establecemos una semilla para que los resultados
# sean reproducibles cada vez que se ejecute el código
np.random.seed(42)


# GENERACIÓN DE VARIABLES

# Generamos una secuencia de números del 1 al 500
# que representan los días analizados
dias = np.arange(1, 501)

# Generamos la producción diaria de forma aleatoria
# entre 500 y 1500 unidades producidas
produccion_diaria = np.random.randint(
    500,
    1500,
    500
)

# Generamos las horas de operación diarias
# entre 8 y 24 horas
horas_operacion = np.random.randint(
    8,
    24,
    500
)

# Generamos temperaturas ambiente utilizando
# una distribución normal con media de 28°C
# y desviación estándar de 5°C
temperatura_ambiente = np.random.normal(
    28,
    5,
    500
)

# CÁLCULO DEL CONSUMO ENERGÉTICO


# Simulamos el consumo energético considerando:
# producción diaria,
# horas de operación,
# temperatura ambiente
# y una variación aleatoria

consumo_energetico = (

    # Influencia de la producción
    produccion_diaria * 0.6

    # Influencia de las horas de operación
    + horas_operacion * 30

    # Influencia de la temperatura ambiente
    + temperatura_ambiente * 10

    # Variación aleatoria para simular condiciones reales
    + np.random.normal(0, 50, 500)

)


# CREACIÓN DEL DATAFRAME

# Construimos el DataFrame respetando
# el orden solicitado en el taller

df = pd.DataFrame({

    # Día de operación
    "Dia": dias,

    # Producción diaria
    "Produccion_Diaria": produccion_diaria,

    # Horas de funcionamiento
    "Horas_Operacion": horas_operacion,

    # Consumo energético
    "Consumo_Energetico": consumo_energetico,

    # Temperatura ambiente
    "Temperatura_Ambiente": temperatura_ambiente

})

# VISUALIZACIÓN DE LOS DATOS


# Mostramos las primeras cinco filas
# para verificar que los datos fueron creados correctamente
display(df.head())
```

#**Interpretación**

La tabla muestra las primeras cinco filas del conjunto de datos generado. Se observa que cuando aumentan la producción diaria y las horas de operación, el consumo energético también tiende a incrementarse. Por ejemplo, el día 3 presenta una producción y horas de operación más altas, lo que genera un consumo energético mayor que otros días.

Estos resultados indican que los datos fueron generados correctamente y reflejan un comportamiento coherente con un entorno industrial real, permitiendo realizar análisis posteriores de manera confiable.


#**2. ANÁLISIS ESTADÍSTICO**

En esta etapa se realizó un análisis estadístico del consumo energético de la industria mediante el cálculo del consumo promedio, máximo y mínimo. Estas medidas permitieron conocer el comportamiento general del consumo eléctrico, identificar los niveles de demanda energética y establecer una base para detectar posibles variaciones que puedan afectar los costos operativos de la empresa.

* **Consumo promedio**

```

# CÁLCULO DEL CONSUMO ENERGÉTICO PROMEDIO

# Calcular el promedio de la columna
# Consumo_Energetico utilizando la función mean de NumPy
consumo_promedio = np.mean(
    df["Consumo_Energetico"]
)

# Mostrar el resultado en pantalla
# indicando el consumo energético promedio
print(
    "Consumo promedio:",
    consumo_promedio
)
```

# **Interpretación**

El consumo promedio representa el comportamiento energético habitual de la industria durante el período analizado. Este indicador constituye una referencia fundamental para evaluar si determinados días presentan consumos superiores o inferiores a lo esperado. Desde una perspectiva gerencial, conocer este valor permite establecer metas de eficiencia energética y realizar comparaciones con futuros periodos de operación.

* **Consumo máximo**






```

# CÁLCULO DEL CONSUMO ENERGÉTICO MÁXIMO

# Obtener el valor máximo de la columna
# Consumo_Energetico utilizando la función max de NumPy
consumo_maximo = np.max(
    df["Consumo_Energetico"]
)

# Mostrar el resultado en pantalla
# indicando el mayor consumo energético registrado
print(
    "Consumo máximo:",
    consumo_maximo
)
```

 # **Interpretación**

Este código calcula el consumo energético máximo registrado en el conjunto de datos. Para ello, se utiliza la función np.max(), la cual examina todos los valores de la columna Consumo_Energetico y devuelve el valor más alto encontrado.

El resultado obtenido representa el mayor nivel de consumo energético registrado durante el período analizado. Este indicador es importante porque permite identificar los momentos de mayor demanda de energía dentro de la empresa. Además, facilita el análisis de picos de consumo, la evaluación de la eficiencia operativa y la detección de posibles situaciones que requieran una gestión más eficiente de los recursos energéticos. Comparar el consumo máximo con otros indicadores estadísticos ayuda a comprender mejor el comportamiento energético general del proceso productivo.

# **Consumo mínimo**

```

# CÁLCULO DEL CONSUMO ENERGÉTICO MÍNIMO


# Obtener el valor mínimo de la columna
# Consumo_Energetico utilizando la función min de NumPy
consumo_minimo = np.min(
    df["Consumo_Energetico"]
)

# Mostrar el resultado en pantalla
# indicando el menor consumo energético registrado
print(
    "Consumo mínimo:",
    consumo_minimo
)
```

# **Interpretación**

Este código calcula el consumo energético mínimo registrado en el conjunto de datos. Para ello, se utiliza la función np.min(), la cual recorre todos los valores de la columna Consumo_Energetico y selecciona el valor más pequeño encontrado.

El resultado obtenido representa el menor nivel de consumo de energía registrado durante el período analizado. Este indicador es importante porque permite identificar los momentos de mayor eficiencia energética o los períodos en los que la demanda de energía fue más baja. **3. GRÁFICO DE SERIE TEMPORAL**

Este gráfico permite visualizar la evolución del consumo energético a lo largo del tiempo, facilitando la identificación de tendencias, variaciones y posibles cambios en el comportamiento del consumo eléctrico de la industria.

```

# SERIE TEMPORAL DEL CONSUMO ENERGÉTICO


# Crear una figura con tamaño de 12 pulgadas de ancho
# y 5 pulgadas de alto para visualizar mejor la evolución
# del consumo energético a lo largo del tiempo
plt.figure(figsize=(12,5))

# Generar un gráfico de líneas donde:
# Eje X = Día
# Eje Y = Consumo Energético
# La línea muestra el comportamiento del consumo
# energético durante el periodo analizado
plt.plot(
    df["Dia"],
    df["Consumo_Energetico"]
)

# Agregar un título descriptivo al gráfico
plt.title("Serie Temporal del Consumo Energético")

# Etiqueta para el eje X
plt.xlabel("Día")

# Etiqueta para el eje Y
plt.ylabel("Consumo (kWh)")

# Agregar una cuadrícula para facilitar la lectura
# e interpretación de los valores
plt.grid()

# Mostrar el gráfico en pantalla
plt.show()
```

# **Interpretación**

La serie temporal permite visualizar el comportamiento del consumo energético a lo largo del tiempo. Se observa que el consumo presenta fluctuaciones constantes, lo que indica que la demanda energética está directamente influenciada por cambios en las condiciones operativas.

Desde una perspectiva de ingeniería, este comportamiento evidencia la necesidad de implementar sistemas de monitoreo continuo que permitan detectar incrementos anormales de consumo antes de que impacten significativamente los costos de producción.

### **4. HISTOGRAMA DEL CONSUMO**

Este histograma permitirá analizar la distribución del consumo energético de la industria, identificando los rangos de consumo más frecuentes y observando la variabilidad de los datos registrados.

```

# HISTOGRAMA DEL CONSUMO ENERGÉTICO


# Crear una figura con tamaño de 8 pulgadas de ancho
# y 5 pulgadas de alto para mejorar la visualización
plt.figure(figsize=(8,5))

# Generar un histograma utilizando la variable
# Consumo_Energetico.
# bins=20 divide los datos en 20 intervalos
# para observar mejor la distribución.
plt.hist(
    df["Consumo_Energetico"],
    bins=20
)

# Agregar un título descriptivo al gráfico
plt.title("Distribución del Consumo Energético")

# Etiqueta para el eje X
plt.xlabel("Consumo")

# Etiqueta para el eje Y
plt.ylabel("Frecuencia")

# Mostrar el gráfico en pantalla
plt.show()
```

#**Interpretación**

El histograma permite identificar cómo se distribuyen los niveles de consumo energético. Si la mayoría de los datos se concentra alrededor del promedio, se puede concluir que el sistema opera bajo condiciones relativamente estables.

Asimismo, la presencia de valores extremos podría indicar eventos atípicos que requieren análisis detallado, ya que podrían estar relacionados con desperdicio energético o fallas en los procesos productivos.

#**GRAFICO DE DISPERSION**

Este gráfico permitirá analizar la relación entre dos variables del estudio, facilitando la identificación de posibles patrones, tendencias o comportamientos que puedan influir en el consumo energético de la industria.

```

# PRODUCCIÓN DIARIA VS CONSUMO ENERGÉTICO


# Crear una figura con tamaño de 8 pulgadas de ancho
# y 5 pulgadas de alto para mejorar la visualización
plt.figure(figsize=(8,5))

# Generar un gráfico de dispersión donde:
# Eje X = Producción Diaria
# Eje Y = Consumo Energético
# Cada punto representa un registro del dataset
plt.scatter(
    df["Produccion_Diaria"],
    df["Consumo_Energetico"]
)

# Agregar título al gráfico
plt.title("Producción Diaria vs Consumo Energético")

# Etiqueta para el eje X
plt.xlabel("Producción Diaria")

# Etiqueta para el eje Y
plt.ylabel("Consumo Energético")

# Mostrar el gráfico en pantalla
plt.show()
```

#**Interpretación**

La gráfica muestra una tendencia ascendente entre ambas variables. Esto evidencia que a medida que aumenta la producción también incrementa el consumo energético.

Sin embargo, resulta importante analizar aquellos puntos que se alejan de la tendencia general, ya que podrían representar días en los cuales se consumió más energía de la necesaria para obtener una determinada producción, reflejando posibles ineficiencias operativas.

##**6. HEATMAP DE CORRELACIÓN**

Este gráfico permite visualizar el nivel de relación existente entre las variables analizadas, facilitando la identificación de posibles asociaciones que puedan influir en el consumo energético de la industria.

```

# MATRIZ DE CORRELACIÓN (HEATMAP)


# Crear una figura de 8 pulgadas de ancho
# y 6 pulgadas de alto para mejorar la visualización
plt.figure(figsize=(8,6))

# Generar un mapa de calor utilizando la matriz
# de correlación de las variables numéricas.
# annot=True muestra los valores de correlación.
# cmap="coolwarm" aplica una escala de colores
# para diferenciar correlaciones positivas y negativas.
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

# Agregar un título descriptivo al gráfico
plt.title("Matriz de Correlación")

# Mostrar el gráfico en pantalla
plt.show()
```

# **Interpretación**

La matriz de correlación permite determinar qué variables tienen mayor influencia sobre el consumo energético.

Si la producción diaria presenta la correlación más alta, se puede afirmar que es el principal factor asociado al gasto eléctrico. Por otro lado, si las horas de operación también muestran una correlación significativa, esto indicaría que la duración de las jornadas productivas impacta considerablemente en el consumo.

Esta información resulta clave para orientar futuras estrategias de optimización.

#**7. PAIRPLOT**

#**Interpretación**

El pairplot proporciona una visión integral de las relaciones existentes entre todas las variables del conjunto de datos.

Este gráfico permite identificar tendencias lineales, agrupamientos y posibles valores atípicos que no son fácilmente observables mediante análisis univariados. Desde el punto de vista de la Inteligencia Artificial, esta herramienta constituye una fase fundamental del análisis exploratorio de datos previo al entrenamiento de modelos predictivos.

#**8. REGPLOT**

Este gráfico permite analizar la relación entre dos variables numéricas mediante una línea de tendencia, facilitando la identificación de posibles asociaciones que puedan influir en el consumo energético de la industria.

```

# GRÁFICO DE REGRESIÓN ENTRE PRODUCCIÓN Y CONSUMO ENERGÉTICO


# Crear una figura con tamaño de 8 pulgadas de ancho
# y 5 pulgadas de alto para mejorar la visualización
plt.figure(figsize=(8,5))

# Generar un gráfico de regresión lineal
# donde:
# x = Producción Diaria
# y = Consumo Energético
# data = DataFrame que contiene los datos
sns.regplot(
    x="Produccion_Diaria",
    y="Consumo_Energetico",
    data=df
)

# Agregar un título descriptivo al gráfico
plt.title(
    "Regresión entre Producción y Consumo"
)

# Mostrar el gráfico en pantalla
plt.show()
```

# **Interpretación**

La línea de regresión confirma matemáticamente la existencia de una relación positiva entre la producción y el consumo energético.

Este resultado permite concluir que el aumento de la actividad productiva tiene un impacto directo sobre la demanda de energía. Sin embargo, la dispersión observada alrededor de la línea indica que existen otros factores que también influyen en el consumo, como la temperatura ambiente o las horas de operación.
