Para el desarrollo de este taller se utilizarán las librerías NumPy, Pandas,Matplotlib y Seaborn, herramientas ampliamente empleadas en proyectos de análisis de datos, ciencia de datos e inteligencia artificial.
```python
# Importación de librerías necesarias para el análisis de datos

import numpy as np
# NumPy: permite trabajar con arreglos multidimensionales y realizar operaciones matemáticas y estadísticas de manera eficiente.

import pandas as pd
# Pandas: facilita la creación y manipulación de estructuras de datos como DataFrames para organizar y analizar información.

import matplotlib.pyplot as plt
# Matplotlib: biblioteca utilizada para crear gráficos y visualizaciones que ayudan a interpretar los datos.

import seaborn as sns
# Seaborn: biblioteca basada en Matplotlib que permite generar gráficos estadísticos más atractivos y fáciles de analizar.
```

**Generación de Datos Sintéticos**

```python

# Generación de datos sintéticos
# Caso 1: Rotación de empleados


n = 500  # Número de empleados simulados

# Generación de la edad de los empleados (22 a 60 años)
edad = np.random.randint(22, 60, n)

# Generación del salario mensual mediante una distribución normal
# Media = $3.500.000 y desviación estándar = $900.000
# Los valores se limitan entre $1.500.000 y $8.000.000
salario = np.random.normal(3500000, 900000, n)
salario = np.clip(salario, 1500000, 8000000)

# Horas extras trabajadas durante el mes
horas_extras = np.random.randint(0, 40, n)

# Antigüedad del empleado en la empresa (1 a 20 años)
antiguedad = np.random.randint(1, 20, n)

# Nivel de satisfacción laboral en una escala de 1 a 11 porque se descarta 1 numero
# El 11 se usa solo como límite y se descarta.
satisfaccion = np.random.randint(1, 11, n)

# Variable objetivo: Renunció


renuncio = []

for i in range(n):

    # Los empleados con baja satisfacción y muchas horas extras
    # presentan una mayor probabilidad de renunciar
    if satisfaccion[i] <= 4 and horas_extras[i] > 20:
        renuncio.append("Sí")

    else:
        # Para los demás casos se asigna una probabilidad
        # del 25% de renunciar y 75% de permanecer
        renuncio.append(
            np.random.choice(
                ["Sí", "No"],
                p=[0.25, 0.75]
            )
        )


# Construcción del DataFrame

datos = pd.DataFrame({
    "Edad": edad,
    "Salario": salario,
    "Horas_Extras": horas_extras,
    "Antiguedad": antiguedad,
    "Satisfaccion": satisfaccion,
    "Renuncio": renuncio
})

# Visualización de las primeras filas del conjunto de datos

pd.options.display.float_format = '{:,.2f}'.format
datos.head()
```

**Interpretación de los datos generados**

En esta etapa se generó un conjunto de datos sintéticos compuesto por 500 empleados de la empresa TechSolutions S.A.S. Para ello se definieron variables que pueden influir en la rotación de personal, tales como la edad, el salario, las horas extras trabajadas, la antigüedad en la empresa y el nivel de satisfacción laboral.

Los valores fueron generados dentro de rangos realistas con el fin de representar situaciones que podrían presentarse en una empresa tecnológica. Además, se creó la variable "Renunció", la cual permite identificar si un empleado decidió abandonar la organización o permanecer en ella. Para su generación se consideró que los trabajadores con bajos niveles de satisfacción y una alta carga de horas extras tienen una mayor probabilidad de renunciar.

Finalmente, los datos fueron organizados en un DataFrame para facilitar su análisis y visualización en las siguientes etapas del estudio. Esto permitirá identificar patrones, tendencias y posibles relaciones entre las variables que intervienen en la rotación de empleados.

**Estadísticas Descriptivas**

```python

# Estadística descriptiva
# Caso 1: Rotación de empleados


# Calcular la edad promedio de los empleados
# Permite conocer la edad representativa de la población analizada
print("Edad promedio:", np.mean(datos["Edad"]))

# Calcular el salario promedio
# Indica el ingreso promedio mensual de los empleados
print("Salario promedio:", np.mean(datos["Salario"]))

# Calcular la desviación estándar del salario
# Mide qué tan dispersos están los salarios respecto al promedio
print("Desviación estándar del salario:", np.std(datos["Salario"]))

# Calcular el porcentaje de empleados que renunciaron
# Se divide el número de empleados que renunciaron entre el total de empleados
porcentaje_renuncia = (
    len(datos[datos["Renuncio"] == "Sí"]) / len(datos)
) * 100

print("Porcentaje de renuncia:", round(porcentaje_renuncia, 2), "%")
```

#**Interpretación de los resultados**

Las estadísticas descriptivas permiten obtener una visión general del comportamiento de los empleados analizados. La edad y el salario promedio ayudan a caracterizar la población de estudio, mientras que la desviación estándar permite evaluar la variabilidad de los salarios dentro de la empresa. Por su parte, el porcentaje de renuncia proporciona una medida del nivel de rotación laboral presente en la organización.

**Gráfico 1 - Histograma de Salarios**

```python

# Visualización: Histograma de salarios


# Crear una figura con un tamaño de 8 pulgadas de ancho y 5 de alto
plt.figure(figsize=(8,5))

# Crear un histograma utilizando la variable Salario
plt.hist(

    # Datos que se van a representar en el histograma
    datos["Salario"],

    # Número de intervalos o barras que tendrá el gráfico
    bins=20,

    # Color de relleno de las barras
    color="skyblue",

    # Color del borde de las barras
    edgecolor="black"
)

# Agregar un título descriptivo al gráfico
plt.title("Distribución de Salarios")

# Agregar etiqueta al eje X
plt.xlabel("Salario")

# Agregar etiqueta al eje Y
plt.ylabel("Frecuencia")

# Mostrar los valores del eje X en formato normal
# evitando la notación científica
plt.ticklabel_format(style='plain', axis='x')

# Agregar una cuadrícula horizontal para facilitar la lectura
plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.7
)

# Mostrar el gráfico en pantalla
plt.show()
```

#**Interpretación del Histograma de Salarios**

El histograma muestra la distribución de los salarios de los 500 empleados simulados de la empresa TechSolutions S.A.S. Se observa que la mayor concentración de empleados se encuentra en salarios cercanos a los 3.500.000 y 4.000.000, lo que indica que estos son los rangos salariales más comunes dentro de la organización.

La forma del gráfico presenta una distribución aproximadamente normal, ya que la mayor cantidad de datos se concentra en el centro y disminuye gradualmente hacia los extremos. Esto significa que existen pocos empleados con salarios muy bajos o muy altos en comparación con la mayoría.

También se puede observar que los salarios se distribuyen entre aproximadamente $1.500.000 y $5.700.000, evidenciando una variabilidad moderada en la remuneración de los trabajadores. En general, la estructura salarial simulada refleja un comportamiento equilibrado y coherente con el de una empresa tecnológica donde la mayoría de los empleados reciben salarios cercanos al promedio establecido.


**Gráfico 2 - Barras de Renuncias**

```python

# Visualización: Gráfico de barras de renuncias


# Contar la cantidad de empleados que renunciaron y los que permanecieron en la empresa
datos["Renuncio"].value_counts().plot(

    kind="bar",                       # Tipo de gráfico: barras
    color=["salmon", "lightgreen"],   # Colores para diferenciar las categorías
    edgecolor="black"                 # Bordes negros para mejorar la visualización
)

# Agregar título al gráfico
plt.title("Cantidad de Empleados que Renunciaron")

# Etiqueta del eje X
plt.xlabel("Renunció")

# Etiqueta del eje Y
plt.ylabel("Cantidad de Empleados")

# Agregar cuadrícula horizontal para facilitar la lectura de las frecuencias
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
```

#**Interpretación del Gráfico de Barras**

El gráfico muestra que de los 500 empleados analizados, aproximadamente 309 permanecieron en la empresa, mientras que cerca de 191 decidieron renunciar. Esto indica que la mayoría de los trabajadores continúan vinculados a la organización, aunque existe una proporción importante de empleados que abandonan la empresa.

La diferencia entre ambas categorías sugiere que la empresa presenta un nivel de rotación moderado, por lo que resulta necesario analizar con mayor detalle las variables relacionadas con la satisfacción laboral, las horas extras y los salarios para identificar posibles factores asociados a las renuncias.

Estos resultados sirven como punto de partida para comprender el comportamiento de los empleados y establecer estrategias orientadas a mejorar la retención del talento humano dentro de la organización.


**Gráfico 3 - Dispersión Salario vs Satisfacción**

```python

# Visualización: Relación entre salario y satisfacción laboral


# Crear una figura con tamaño de 8x5 pulgadas
plt.figure(figsize=(8,5))

# Crear gráfico de dispersión
plt.scatter(
    datos["Salario"],          # Variable del eje X
    datos["Satisfaccion"],     # Variable del eje Y

    # Colorear los puntos según la variable Renunció
    c=(datos["Renuncio"] == "Sí"),

    # Escala de colores
    cmap="coolwarm",

    # Nivel de transparencia de los puntos
    alpha=0.6,

    # Color del borde de cada punto
    edgecolor="black"
)

# Agregar título al gráfico
plt.title("Relación entre Salario y Satisfacción Laboral")

# Etiqueta del eje X
plt.xlabel("Salario")

# Etiqueta del eje Y
plt.ylabel("Nivel de Satisfacción")

# Evitar la notación científica en el eje X
plt.ticklabel_format(style='plain', axis='x')

# Agregar cuadrícula para facilitar la lectura
plt.grid(True, linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
```

### **Interpretación del Gráfico de Dispersión**

El gráfico de dispersión muestra la relación entre el salario y el nivel de satisfacción laboral de los empleados. Cada punto representa un trabajador de la empresa y su ubicación depende del salario que recibe y de su nivel de satisfacción.

Al observar la distribución de los puntos, no se identifica una relación lineal clara entre ambas variables, ya que empleados con salarios similares presentan diferentes niveles de satisfacción. Esto indica que el salario no es el único factor que influye en la percepción laboral de los trabajadores.

También se observa que los empleados que renunciaron se encuentran distribuidos en distintos rangos salariales, aunque existe una mayor presencia de renuncias en los niveles bajos de satisfacción. Esto sugiere que la satisfacción laboral podría tener una influencia más importante en la decisión de abandonar la empresa que el salario por sí solo.

En general, el gráfico evidencia que la rotación de personal puede estar asociada a múltiples factores, por lo que es necesario analizar otras variables como las horas extras y la antigüedad para comprender mejor el comportamiento de los empleados.


**Gráfico 4 - Heatmap de Correlación**

```python

# Visualización: Matriz de correlación


# Seleccionar únicamente las variables numéricas del conjunto de datos
# y calcular la correlación entre ellas
correlacion = datos.select_dtypes(include=np.number).corr()

# Crear una figura de 8x6 pulgadas
plt.figure(figsize=(8,6))

# Construir el mapa de calor de correlaciones
sns.heatmap(

    correlacion,      # Matriz de correlación calculada

    annot=True,       # Mostrar el valor numérico en cada celda

    cmap="coolwarm",  # Escala de colores para identificar correlaciones

    fmt=".2f",        # Mostrar los valores con dos decimales

    linewidths=0.5    # Agregar separación entre las celdas
)

# Agregar título al gráfico
plt.title("Matriz de Correlación entre Variables")

# Mostrar el gráfico
plt.show()
```

### **Interpretación de la Matriz de Correlación**

La matriz de correlación permite analizar el grado de relación existente entre las variables numéricas del conjunto de datos. Al observar los resultados, se evidencia que la mayoría de los coeficientes de correlación se encuentran muy cerca de cero, lo que indica relaciones débiles o prácticamente inexistentes entre las variables analizadas.

La correlación más alta se presenta entre salario y satisfacción laboral (0.10), aunque este valor sigue siendo muy bajo para considerar una relación significativa. De igual forma, la edad presenta una correlación negativa débil con el salario (-0.08), mientras que las demás variables muestran valores cercanos a cero.

Estos resultados sugieren que ninguna variable por sí sola explica el comportamiento de las demás, por lo que la decisión de renunciar podría depender de la combinación de varios factores y no únicamente de una característica específica. Este comportamiento es común en problemas relacionados con la gestión del talento humano, donde intervienen múltiples aspectos laborales y personales.


**Gráfico 5 - Boxplot de Salario**

```python

# Visualización: Boxplot de salarios


# Crear una figura con tamaño de 8x5 pulgadas
plt.figure(figsize=(8,5))

# Generar el diagrama de caja para la variable Salario
sns.boxplot(
    x=datos["Salario"],
    color="lightblue"
)

# Agregar título al gráfico
plt.title("Distribución de Salarios mediante Boxplot")

# Etiqueta del eje X
plt.xlabel("Salario")

# Evitar que los valores aparezcan en notación científica
plt.ticklabel_format(style='plain', axis='x')

# Mostrar el gráfico
plt.show()
```

**Interpretación del Boxplot de Salarios**

El boxplot permite observar la distribución de los salarios de los empleados analizados. La mediana se ubica aproximadamente en $3.500.000, lo que indica que el 50 % de los empleados percibe salarios superiores a este valor y el otro 50 % salarios inferiores.

Asimismo, el rango intercuartílico, que contiene el 50 % central de los datos, se encuentra aproximadamente entre 2.900.000 y 4.000.000, evidenciando una dispersión moderada de los salarios dentro de la empresa. Los valores mínimos y máximos se sitúan alrededor de 1.500.000 y 5.700.000, respectivamente.

No se observan valores atípicos (outliers), lo que sugiere que los salarios se mantienen dentro de rangos esperados para la simulación realizada. En general, la distribución presenta un comportamiento relativamente equilibrado, aunque se aprecia una ligera dispersión hacia los salarios más altos.

**Gráfico 6 - Pairplot**

```python

# Visualización: Pairplot de variables


# Crear un pairplot para analizar simultáneamente las relaciones
# entre todas las variables numéricas del conjunto de datos
sns.pairplot(

    datos,              # DataFrame que contiene la información de los empleados

    hue="Renuncio",     # Diferenciar los empleados según hayan renunciado o no

    diag_kind="kde",    # Mostrar curvas de densidad en la diagonal principal

    palette="coolwarm"  # Paleta de colores utilizada para las categorías
)

# Mostrar el gráfico generado
plt.show()
```

**Interpretación de la matriz de dispersión (Pairplot)**


El pairplot permite observar la distribución de las variables y las relaciones existentes entre ellas, diferenciando a los empleados que renunciaron de aquellos que permanecieron en la empresa.

A partir de la gráfica se observa que la edad, el salario y la antigüedad presentan comportamientos similares en ambos grupos, por lo que no parecen tener una influencia significativa en la decisión de renunciar. Por otro lado, las horas extras y el nivel de satisfacción muestran diferencias más evidentes, ya que los empleados que renunciaron tienden a presentar mayores cargas de trabajo y menores niveles de satisfacción laboral.

En general, no se identifican correlaciones lineales fuertes entre las variables analizadas; sin embargo, la satisfacción laboral y las horas extras parecen ser los factores más relacionados con la rotación de personal dentro de la empresa.

Importación de librerías

```python
# Importamos la librería NumPy para trabajar con números y estadísticas
import numpy as np

# Importamos Pandas para organizar los datos en forma de tabla
import pandas as pd

# Importamos Matplotlib para crear gráficos básicos
import matplotlib.pyplot as plt

# Importamos Seaborn para gráficos estadísticos avanzados
import seaborn as sns

# Configuramos un estilo visual agradable para los gráficos
sns.set(style="whitegrid")

# Fijamos una semilla para que los resultados sean reproducibles
np.random.seed(42)
```
